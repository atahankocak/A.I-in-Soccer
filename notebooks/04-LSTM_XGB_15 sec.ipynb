```python
# read in 1st period data from PlayerTek tracking system coordinate csv file
period_1 = pd.read_csv("../../csv_files/AI_in_Soccer/LSTM_train.csv", index_col=0)
```

```python
# process the data for training
def preprocess_train(ds, sec, input_count):
    
    # set an upper limit to build input-output datasets to merge. 
    #This enables the most amount of data to be used in training
    upper_limit = len(ds) - (len(ds) % (sec)) - sec
    
    # initilize inputs and outputs from the data
    input_ds = ds[['x_player', 'y_player']]
    input_ds = input_ds.iloc[:upper_limit]    
    output_ds = ds[['x_player', 'y_player']]
    output_ds = output_ds.iloc[sec:upper_limit+sec]  
    
    # reshape into array
    input_ds = input_ds.values
    output_ds = output_ds.values
    
    # group them
    """
    Reshape into groups  for many-to-many sequence
    to form groups of 2 in groups of the count of seconds stated.
    In other words, setting the dataset to train the model 
    with the coordiantes in the stated (sec) portions of time.
    """
    groups = int(input_ds.shape[0] / sec)
    input_ds = np.array(input_ds).reshape(groups, sec, input_count)
    output_ds = np.array(output_ds).reshape(groups, sec, input_count)
    
    return input_ds, output_ds
```

```python
# model training - Bidirectional LSTM
sec = 15
input_count = 2
output_count = 2
input_ds, output_ds = preprocess_train(period_1, sec, 2)
units = 100
activation = "relu"
optimizer = "adam" 
loss = "mse"
metric = "mae"
epochs = 500
split = 0.2
batch_size = 5
verbose = 1
    
# bidirectional LSTM
model = Sequential()
model.add(Bidirectional(LSTM(units, activation=activation, input_shape=(sec,input_count))))
model.add(RepeatVector(sec))
model.add(Bidirectional(LSTM(units, activation=activation, return_sequences=True)))
model.add(TimeDistributed(Dense(output_count)))
model.compile(optimizer=optimizer, loss=loss, metrics=[metric])

history = model.fit(input_ds, output_ds, 
                    epochs=epochs, 
                    validation_split=split, 
                    verbose=verbose, 
                    batch_size=batch_size)
```

```python
# create a scaled results final dataframe to visualize
def predictor(ds, sec, minute, position):
    
    # test dataframes for input and actuals
    n = sec
    m = int(minute * 60)

    input_test = ds[['x_player', 'y_player']]
    input_test = input_test.iloc[m:n+m]      # add 1 to account for end coordinates
    input_test.rename(columns={"x_player": "x_input", "y_player": "y_input"}, inplace=True)
    
    # actual output - This section is for comparison purposes - for development, not production
    output_test = ds[['x_player', 'y_player']]
    output_test = output_test.iloc[n+m-1:n+m+n-1]
    output_test.rename(columns={"x_player": "x_actual", "y_player": "y_actual"}, inplace=True)
    
    # reset index
    input_test = input_test.reset_index(drop=True)
    output_test = output_test.reset_index(drop=True)   
    
    # concat
    dataset = pd.concat([input_test, output_test], axis=1)
    
    # group
    group = int(input_test.shape[0] / n)
    test = np.array(input_test).reshape(group, n, 2)
    
    # predict
    predicted = np.clip(model.predict(test), 0, 100)
    
    # add to the dataframe
        # list the predictions
    pred_list = []
    for i in range (len(predicted)):
        for j in range(n):
            pred_list.append(predicted[i][j])
        # conver to dataframe
    pred_df = pd.DataFrame(pred_list, columns=['x_pred', 'y_pred'])
    
        # add to dataset to be used in viz later
    dataset = pd.concat([dataset, pred_df], axis=1)
    
    # add end coordinates (simply the next coordinate)
    dataset['x_pred_end'] = dataset['x_pred'].shift(-1)
    dataset['y_pred_end'] = dataset['y_pred'].shift(-1)  
    
    # now time to predict events
    xgb_pred = []
    for i in range(0, dataset.shape[0]-1):
        x = [dataset.x_pred[i], dataset.y_pred[i], dataset.y_pred_end[i], dataset.y_pred_end[i], position]
        xgb_pred.append(xgb.predict(x))

    df_xgb = pd.DataFrame(xgb_pred, columns=['event_pred'])

    dataset = pd.concat((dataset,df_xgb),axis=1)

    # scale
    dataset['x_pred_scaled'] = dataset['x_pred'].map(lambda x: x * 120 / 100)
    dataset['y_pred_scaled'] = dataset['y_pred'].map(lambda x: x * 80 / 100)
    dataset['x_pred_end_scaled'] = dataset['x_pred_end'].map(lambda x: x * 120 / 100)
    dataset['y_pred_end_scaled'] = dataset['y_pred_end'].map(lambda x: x * 80 / 100)
    dataset['x_actual_scaled'] = dataset['x_actual'].map(lambda x: x * 120 / 100)
    dataset['y_actual_scaled'] = dataset['y_actual'].map(lambda x: x * 80 / 100)
    dataset['x_input_scaled'] = dataset['x_input'].map(lambda x: x * 120 / 100)
    dataset['y_input_scaled'] = dataset['y_input'].map(lambda x: x * 80 / 100)
    
    # dominant event
    dominant_event = dataset.event_pred.value_counts().index.tolist()[0]
    
    # visualize - product
    football_pitch_2(1)
    ax = plt.scatter(dataset['x_pred_scaled'], dataset['y_pred_scaled'], c='r', edgecolors="k", zorder=12, marker ='o', linewidth=.5)
    ax2=plt.scatter(dataset['x_input_scaled'], dataset['y_input_scaled'], marker='o', c='y', edgecolors="w", linewidth=0.25, zorder=12)
    plt.title(f'Dominant event is {dominant_event}', fontsize=20)
    plt.legend([ax2, ax],['Past 15 Sec', 'Next 15 sec'], loc='upper right', prop={'size': 16})
    plt.show()
    
    
    return dominant_event, plt.show()
    
```
